```python
import sqlite3

# create an in-memory database
conn = sqlite3.connect(":memory:")
c = conn.cursor()

# create a sample table
c.execute("""
CREATE TABLE people (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")

# insert fake data
c.executemany(
    "INSERT INTO people (name, age) VALUES (?, ?)",
    [("Alice", 30), ("Bob", 25), ("Charlie", 35)]
)

conn.commit()

# run a query
for row in c.execute("SELECT * FROM people WHERE age > 28"):
    print(row)

conn.close()
```
